```python
# Task 1.5 
time_array = []
n_array = [i for i in range(1,n+1)]

for i in range(1,n+1):
  mean_array = []
  for j in range(5):
    v = [random.randint(1,99) for m in range(i)]
    start_time = time.time()
    c = 0
    while c < i - 1:
      p = 0
      while p < i - 1 - c:
        if v[p] > v[p+1]:
          v[p], v[p+1] = v[p+1], v[p]
        p += 1
      c += 1
    mean_array.append(time.time()-start_time)
  time_array.append(mean(mean_array))


const = time_array[-1]/n_array[-1] ** 2
y = [const * i ** 2 for i in n_array]

plt.plot(n_array,time_array, label='experiment')
plt.plot(n_array, y, label='O(n^2)')
plt.legend()
plt.xlabel('n')
plt.ylabel('time,sec')
plt.grid()
plt.show()
```

```python
# Task 1.7
# Timsort is python's built-in sort
time_array = []
n_array = [i for i in range(1,n+1)]

minrun = 32

def InsSort(arr,start,end):    
    for i in range(start+1,end+1):
        elem = arr[i]
        j = i-1
        while j>=start and elem<arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = elem
    return arr
def merge(arr,start,mid,end):
    if mid==end:
        return arr
    first = arr[start:mid+1]
    last = arr[mid+1:end+1]
    len1 = mid-start+1
    len2 = end-mid
    ind1 = 0
    ind2 = 0
    ind  = start
     
    while ind1<len1 and ind2<len2:
        if first[ind1]<last[ind2]:
            arr[ind] = first[ind1]
            ind1 += 1
        else:
            arr[ind] = last[ind2]
            ind2 += 1
        ind += 1
     
    while ind1<len1:
        arr[ind] = first[ind1]
        ind1 += 1
        ind += 1
              
    while ind2<len2:
        arr[ind] = last[ind2]
        ind2 += 1
        ind += 1   
              
    return arr
            
def TimSort(arr):
    n = len(arr)
    
    for start in range(0,n,minrun):
        end = min(start+minrun-1,n-1)
        arr = InsSort(arr,start,end)
        
    curr_size = minrun
    while curr_size<n:    
        for start in range(0,n,curr_size*2):
            mid = min(n-1,start+curr_size-1)
            end = min(n-1,mid+curr_size)
            arr = merge(arr,start,mid,end)
        curr_size *= 2
    return arr

for i in range(1,n+1):
  mean_array = []
  for j in range(5):
    v = [random.randint(1,99) for m in range(i)]
    start_time = time.time()
    v = TimSort(v)
    mean_array.append(time.time()-start_time)
  time_array.append(mean(mean_array))

const = time_array[-1]/(n_array[-1]*math.log10(n_array[-1]))
y = [i*math.log10(i)*const for i in n_array]

plt.plot(n_array,time_array, label='experiment')
plt.plot(n_array, y, label='O(nlogn)')
plt.legend()
plt.xlabel('n')
plt.ylabel('time')
plt.grid()
plt.show()
```

```python
# Task 2
time_array = []
n = 150
n_array = [i for i in range(1,n+1)]

for i in range(n):
  mean_array = []
  for j in range(5):
    C = np.zeros((i,i))
    A = np.random.randint(1,99, size=(i,i))
    B = np.random.randint(1,99, size=(i,i))
    start_time = time.time()
    for k in range(i):
      for m in range(i):
        for p in range(i):
          C[k][m] += A[k][p]*B[p][m]
    mean_array.append(time.time()-start_time)
  time_array.append(mean(mean_array))

const = time_array[-1]/n_array[-1]**3
y = [const * i ** 3 for i in n_array]

plt.plot(n_array,time_array, label='experiment', color='blue')
plt.scatter(n_array, y, label='n^3', s=3, color='red')

plt.legend()
plt.xlabel('n')
plt.ylabel('time,sec')
plt.grid()
plt.show()
```
